```python
import os 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def link_collector(url):
    response = requests.get(url)
    html = response.content
    response.close()
    soup = BeautifulSoup(html, 'lxml')
    main_part = soup.find('div', attrs={'id':'content'})
    return_list = []
    for tag in main_part.findAll('a'):
        link = tag.get('href')
        try:
            
            if link.split('adatlapjai')[1][0] == '&':
                pass
            elif link.split('adatlapjai/')[1][1] == '/':
                pass
            elif link.split('adatlapjai/')[1] == 'num/':
                pass
            elif link.split('adatlapjai')[1][1] == '&':
                pass
            else:
                return_list.append(link)
        except:
            pass
    return return_list
```

```python
link_collector('http://magyarbrands.hu/magyar-markak/magyar-markak-adatlapjai/&start=0')
```

```python
link_list = []
for x in range(40):
    for link in link_collector('http://magyarbrands.hu/magyar-markak/magyar-markak-adatlapjai/&start={}'.format(10*x)):
        link_list.append(link)
```

```python
def page_scraper(url):
    response = requests.get(url)
    html = response.content
    response.close()
    name, website, email, phone, address, description = None, None, None, None, None, None
    soup = BeautifulSoup(html, 'lxml')
    middle_soup = soup.find('div', attrs={'class':'content articles'})
    name = middle_soup.find('h1').text
    description = ''
    for paragraph in middle_soup.findAll('p'):
        if "Weboldal:" in paragraph.text:
            try:
                website = paragraph.find('a').get('href')
            except:
                pass
        elif "Cm:" in paragraph.text:
            address = paragraph.text.split('Cm:')[1].split('Telefon')[0].replace('\r\n', '')
            if 'Telfon' in paragraph.text:
                phone = paragraph.text.split('Cm:')[1].split('Telefon: ')[1]
        elif "E-mail: " in paragraph.text:
            email = paragraph.text.split('E-mail: ')[1].replace('\xa0', '')
        elif "bemutatsa s trtnete" in paragraph.text:
            pass
        else:
            description += paragraph.text
    return_list = [name, website, email, phone, address, description]
    return(return_list)
```

```python
page_scraper('http://magyarbrands.hu/magyar-markak/magyar-markak-adatlapjai/volanbusz/')
```

```python
list_of_lists = []
for link in link_list:
    print(link)
    list_of_lists.append(page_scraper(link))
```

```python
df = pd.DataFrame(list_of_lists, columns=['Name', 'Url', 'Email', 'Phone', 'Address', 'Description'])
```
