```python
import os 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
main_url = 'http://www.magyarvallalkozasok.hu/'
output_path = os.path.join(os.path.dirname('__file__'), '..',) +'/output/'
```

```python
response = requests.get(main_url)
html = response.content
response.close()
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
middle = soup.find('div', attrs={'id':'middle'})
```

```python
tag_list = []
for tag in middle.findAll('a'):
    text = tag.text
    link = tag.get('href')
    tag_list.append([text, link])
```

```python
tag_list
```

```python
def page_scraper(url):
    response = requests.get(url)
    content = response.content
    response.close()
    soup = BeautifulSoup(content, 'lxml')
    middle_soup = soup.find('div', attrs={'id':'middle'})

    company_list = []
    for tag in middle_soup.findAll('h1'):
        address = tag.text
        company_name = tag.find('a').text
        company_link = tag.find('a').get('href')
        address = address.replace(company_name, '')
        company_list.append([company_name, address])
    for x in range(len(company_list)):
        company_list[x].append(middle_soup.findAll('a')[x*2+1].get('href'))
    print(middle_soup.find('h1'))
    return company_list
```

```python
list_of_lists = []
for tag in tag_list:
    sector = tag[0]
    url = 'http://www.magyarvallalkozasok.hu/' + tag[1]
    data = page_scraper(url)
    for x in data:
        x.append(sector)
    list_of_lists.append(data)
```

```python
list_of_lists = page_scraper('http://www.magyarvallalkozasok.hu/index.php?cid=62')
```

```python
final_list = []
for sector in list_of_lists:
    for company in sector:
        final_list.append(company)
```

```python
df = pd.DataFrame(final_list, columns=['Name', 'Address', 'Url', 'Sector'])
```
